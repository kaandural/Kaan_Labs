```python
import requests 
from bs4 import BeautifulSoup 
import pandas as pd
import numpy as np
import re
```

```python
url = "https://www.billboard.com/charts/hot-100/"
```

```python
response = requests.get(url)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup.select("li>h3")
```

```python
titles = []
for i in soup.select("li > h3.c-title"):
    titles.append(i.get_text("|", strip = True).lower())
titles
```

```python
artists = []
for i in soup.select("li > span.c-label.a-no-trucate"):
    artists.append(i.get_text("|", strip = True))
len(artists)
```

```python
df = pd.DataFrame({'artists': artists, 'songs': titles})
```

```python
df
```

```python
user_song = input("Enter a song: ")
user_song = user_song.lower()

if user_song in df['songs'].values:
    recommended_song = df.sample()['songs'].values[0]
    print(f"We recommend you listen to: {recommended_song}")
else:
    print("Sorry, we don't have any song recommendations for you.")
```

```python
import random

# Assuming you already have the 'top_100' DataFrame with 'Title' column converted to lowercase

def suggest_title(input_title, df):
    # Convert the input_title to lowercase to ensure case-insensitivity
    input_title = input_title.lower()
    # Check if the input_title is in the 'Title' column of the DataFrame
    if input_title in df['songs'].values:
        # Get all titles excluding the input_title itself
        suggestions = df.loc[df['songs'].str.lower() != input_title, 'songs'].values
        # Return one suggestion randomly
        suggested_title = random.choice(suggestions)
        return suggested_title
    else:
        return "This song is not trending. You are out of touch."

# Get user input
user_input = input("Enter a song title: ")

# Get and print the suggestion
suggestion = suggest_title(user_input, df)
print(suggestion)
```

```python
user_song = input("Enter a song: ")
user_song = user_song.lower()

if user_song in df['songs'].values:
    recommended_song = df.sample()['songs'].values[0]
    recommended_song = recommended_song.capitalize()  
    print(f"We recommend you listen to: {recommended_song}")
else:
    print("Sorry, we don't have any song recommendations for you.")
```
